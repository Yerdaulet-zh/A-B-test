### An example to consider
To illustrate the concepts applied in hypothesis testing about a population, let's consider this example. A ceritan arbitary type of a vaccine for cold is kwown to be effective for only 25% after a period of 2 years. Researchers decided to test more expensive vaccine to determine if it is superior in providing protection against the same virus for the same period of time as former. 20 people were selected unbiasedly and inoculated. (In practise the number of participants receiving the vaccine would be serveral thousand. For the sake of simplicity in understaning the matrials, 20 people were chosen.) The alternating vaccine surpass the 2 years of period with 8 participans without contracting the virus, the expancive vaccine will be considered superior to the one presently in use. The requirement for 8 people is kind of arbitary but it appears reasonable modest gain over the 5 people who could be expected to receive protection if the 20 people had beed inoculated with the current vaccine. The null hypothesis will say that the current and more expensive one are equally effective, do difference in probabilities. The alternative hypothesis is that the expensive vaccine is superior. This is same to testing hypothesis that the binomial parameter for the probability of a success on a given trial is p =1/4 against the alternative p>1/4. This is usually written as follows: 

$H_0$: p = 0.25

$H_1$: p > 0.25

### The Test Statistic

The **test statistics** on which we base the decision is X, the numver of individuals in pur test group who recieve protection from new vaccine for a period of at least 2 years. The possoble values of X, from 0 to 20, are devided into two groups:
Those numbers less than or equal to 8 and those greater than 8. All possbile values which greater than 8 are considerend to be in **critical region**. The last number that we observe passing into the critical region is called **critical value**. In the illustration, the critical value is 8. Thus, if x > 8, we reject $H_0$ in favor of the alternative hypothesis $H_1$. if x <= 8, we fail to reject $H_0$. This decision criterion can be visualised as follows:

### The Probability of a Type I Error

The decision procedure that described above could lead to either of two wrong conclutions. For instance, the expencive vaccine may be no better than the one now in use ($H_0$ is true) and yet, in this paticular randomly selected group of individuals, more than 8 surpass the 2-year period without contracting the virus. We would be committing an error by rejection $H_0$ in favor of $H_1$ when, in fact, $H_0$ is true. Such an error is called a **type I error**. 

> **<em>Definition: Rejection of the null hypothesis when it is true is called a type I error.</em>**

The second type of error is committed if 8 or fewer of the group surpass the 2-year period successfully and we are unable to colculde that the vaccine is better when it actually is better ($H_1$ true.). This, in this case, we fail to reject $H_0$ when in fact $H_0$ is false. This is called a **type II error.**

> **Definition: Nonrejection of the null hypothesis when it is false is called a type II error.**

In hypothesis testing there are four possible solutions that determine where a decision is correct or in error.

The probability of commiting a **type I error**, also called the **level of significanse**, is denoted by the Greek letter $\alpha$. In the illustration, a **type I error will occur** when more tham 8 individuals inoculated with the expencive vaccine surpass the 2-year old period without getting sick with the virus of cold and researchers conclude that the **new vaccine** is **better** when it is **actually equivalent to the one in use**. 


Let the **significance level** be defined as the probability of a **Type I error**:

$$
\alpha = P(\text{Type I error}) = P(X > 8 \mid p = \tfrac{1}{4})
$$

This can be calculated using the **cumulative binomial distribution**:

$$
\alpha = \sum_{x=9}^{20} \binom{20}{x} \left(\tfrac{1}{4}\right)^x \left(\tfrac{3}{4}\right)^{20 - x}
$$

Alternatively, using the **complement rule**:

$$
\alpha = 1 - \sum_{x=0}^{8} \binom{20}{x} \left(\tfrac{1}{4}\right)^x \left(\tfrac{3}{4}\right)^{20 - x}
$$

Given:

$$
\sum_{x=0}^{8} b(x; 20, \tfrac{1}{4}) = 0.9591
$$

Then:

$$
\alpha = 1 - 0.9591 = 0.0409
$$

<em>The complement rule in probability refers to the idea that the probability of an event not happening is equal to 1 minus the probability that it does happen.</em>

--- 

The null hypothesis p=1/4 is being tested at the $\alpha$=0.0409 level of significance. Sometimes, **the level of significance** is referred to as the **size of the test**. A **critical region** of size 0.0409 is quite small, meaning that the **probability** of committing a **Type I error** is low. Consequently, it would be unusual for more than 8 individuals to remain healthy from the virus over a 2-year period if the proposed vaccine were truly equivalent to the one currently on the market.

### The Probability of a Type II Error

To calculate the probability of a **type II error** denoted by $\beta$ is impossible unless we have a specific alternative hypothesis. If we test the $H_0$: p = 1/4 against $H_1$: 1/2, then we are able to compute the probability of not rejecting $H_0$ when it is false. We simply calculate a comulative probability of obtaining 8 or fewer in the group that surpass the specified period of time whem p=1/2. 

$$
\beta = P(\text{Type II error}) = P\left( X \leq 8 \text{ when } p = \tfrac{1}{2} \right)
$$

$$
= \sum_{x=0}^{8} b\left(x; 20, \tfrac{1}{2} \right) = 0.2517.
$$

This is a rather high probability, indicating a test procedure in which it is quite
likely that we shall **reject the new vaccine** when, **in fact**, it is **superior to what is
now in use**. Ideally, we like to use a test procedure for which the type I and type
II error probabilities are both small.

It is possible, when director does not care much about detecting small differences that don’t justify the cost of switching to a more expensive vaccine. The only time he wishes to guard against the Type II error is when the true value of p is at least 0.7.


$$
\beta = P(\text{Type II error}) = P\left( X \leq 8 \text{ when } p = 0.7 \right)
$$

$$
= \sum_{x=0}^{8} b\left(x; 20, 0.7 \right) = 0.0051.
$$

With such a small probability of committing a type II error, it is extremely unlikely
that the new vaccine would be rejected when it was 70% effective after a period of
2 years. As the alternative hypothesis approaches unity, the value of $\beta$ diminishes
to zero.

### The Role of α, β, and Sample Size

Let's assume that the we are unwilling to commit a type II error when the $H_1$: p = 1/2 is true, even though we have found the probability of such an error to be $\beta$ = 0.2517. It is possible to reduce $\beta$ by increasing the sample size of each critical region. For instance, consider what happens to the values of $\alpha$ and $\beta$ when we change the critical value to 7 so that all scores greater than 7 fall in the critical region and those less than or equal to 7 fall in the nonrejection region. Now, in testing p = 1/4 against the $H_1$ that p = 1/2, we find that:

$$
\alpha = \sum_{x=8}^{20} b\left(x; 20, \tfrac{1}{4} \right)
= 1 - \sum_{x=0}^{7} b\left(x; 20, \tfrac{1}{4} \right)
= 1 - 0.8982 = 0.1018
$$

and 

$$
\beta = \sum_{x=0}^{7} b\left(x; 20, \tfrac{1}{2} \right) = 0.1316.
$$


As a result of new decision procedure, we could reduce the probability of commiting a type II error at the expense of increasing the probability of committing a type I error. For the fixed sample size, a decrease in the probability of one error will result in an increase in the probability of the other error. Fortunatelly, **the probability of commiting both types of error can be reduced by increasing the sample size**. Consider the same problem using a random sample of 100 individuals. If more than 36 of the group surpass the 2-year period, we reject tje null hypothesis that p=1/4 and accept the alternative hypothesis that p > 1/4. The critical value is now 36. All possible scores above 36 constitute the critical region, and all possible scores less than or equal to 36 fall in the acceptance region. 

To determine the probability of committing a Type I error, we shall use the
normal curve approximation with
$$
\mu = np \quad \text{and} \quad \sigma = \sqrt{np(1 - p)},
$$

where \( n = 20 \) and \( p = \tfrac{1}{4} \) under the null hypothesis.
Then we compute
$$
P(X \geq 9) \approx P\left(Z \geq \frac{9 - 0.5 - \mu}{\sigma} \right)
$$
using the standard normal distribution.
