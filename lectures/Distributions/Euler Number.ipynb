Historically the origin of **$e$** start from $17^{th}$ centry, and it originated from studying **continuous growth** and **compounding**.

### 1. Compound Interest Problem (Jacob Bernoulli, 1683).

Bernoulli was studying the formula for compound interest:
$$
\left(1 + \frac{1}{n}\right)^n
$$

**as n grows large**. He found that as **compounding frequency increases without bound**, the **expression approaches a constant value** - which turned out to be **$e$**. 

I provide a visual representation of approximation below. 

```python
import math
import numpy as np
import matplotlib.pyplot as plt

n_values = [1, 2, 4, 8, 16, 32, 64]
approximations = [(1 + 1/n)**n for n in n_values]

plt.figure(figsize=(8, 5))
plt.plot(n_values, approximations, label=r'$(1 + 1/n)^n$ approximation', color='blue')
plt.axhline(math.e, color='red', linestyle='--', label=r'True $e$')

plt.xlabel('n (compounding frequency)')
plt.ylabel('Value')
plt.title('Bernoulli’s Approximation of $e$')
plt.legend()
plt.grid(True)
plt.show()
```

### 2. Logarithms and John Napier (1614)

**Napier introduced logarithms to simplify calculations**. While his **logarithm tables weren’t directly based on e**, his work **led mathematicians** to **explore natural logarithms**, whose base is **$e$**. 

### 3. Analysis and Calculus (Leonhard Euler, 1727–1748)

**Euler formalized $e$ as the base of the natural logarithm** and showed many **equivalent definitions**:

- As a limit: $\lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^{n}$
- As Taylor series: $e = \sum^{\infty}_{k=0}\frac{1}{k!}$
- As the unique base where the derivative of $e^x$ equals $e^x$.

So, $e$ emerged first from **financial mathematics**, was hinted at in **logarithms**, and was later defined rigorously in **calculus** by Euler.

---
#### Euler Definitions:

1. As I demonstrated graph above regarding compounding interest grows, as the value of n grows infinitelly, it will approximate a constant irrational value $e$. 

2. Euler defined the value of $e$ can be **approximated by infinite sum of polynomials Taylor Series**.

3. To proof it, let's take Taylor Series approximation equation:

Approximation:

$e^x = \sum^{\infty}_{n=0}\frac{x^n}{n!}$

Differentiate term-by-term:

$$
\frac{d}{dx}e^x = \frac{d}{dx}(1 + x + \frac{x^2}{2!} + \frac{x^3}{3!} + ... ) = 0 + 1 + \frac{x^2}{2!} + \frac{x^3}{3!} ... = \frac{d}{dx}(1 + x + \frac{x^2}{2!} + \frac{x^3}{3!} + ... = e^x
$$

**So $e^x$ is its own derivative.**
